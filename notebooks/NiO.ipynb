```python
data = np.loadtxt('../src/NiO/nio.DOS.Tetrahedron')
energy = data[:, 0]
dos = data[:, 1]

fig = plt.figure(figsize = (12, 6))
plt.plot(energy, dos, linewidth=0.75, color='red', label="W/O U")

data = np.loadtxt('../src/NiO/nio-U.DOS.Tetrahedron')
energy = data[:, 0]
dos = data[:, 1]

plt.plot(energy, dos, linewidth=0.75, color='green', label="DFT+U")
plt.yticks([])
plt.xlabel('Energy (eV)')
plt.ylabel('DOS')
plt.ylim(0, )
plt.xlim(-5, 5)
plt.legend(frameon=False)
plt.show()
```
